```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
cities_temp = pd.read_csv('Average Temperature of Cities.csv')
```

```python
cities_temp.head()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Year']
```

```python
# Cleaning the wikipedia data so as to have the temperature in Celsius only
for month in months:
    cities_temp[month] = cities_temp[month].apply(lambda x: x.split('\n')[0])
```

```python
# Cleaning the continent column to be lower case
cities_temp['Continent'] = cities_temp['Continent'].apply(lambda x: x.lower())
```

```python
for month in months:
    cities_temp[month] = cities_temp[month].astype(float)
```

```python
cities_temp
```

```python
# Which is the coldest city?
cities_temp[cities_temp['Year'] == cities_temp['Year'].min()]
```

```python
# Which is the hottest city?
cities_temp[cities_temp['Year'] == cities_temp['Year'].max()]
```

```python
# Now we analysze the dataset containing population data
cities_pop = pd.read_csv('cities15000.csv', encoding_errors='replace')
```

```python
cities_pop.head()
```

```python
# Dropping irrelavant columns
cities_pop.drop(['cc2', 'admin2 code'],axis=1, inplace=True)
cities_pop.drop(['admin3 code', 'admin4 code', 'elevation', 'modification date'], axis=1, inplace=True)
cities_pop.drop(['name', 'alternatenames', 'dem'], axis=1, inplace=True)
cities_pop.drop(['geonameid', 'feature class', 'feature code', 'admin1 code', 'timezone'], axis=1, inplace=True)
cities_pop.nunique()
```

```python
cities_pop.head()
```

```python
# Dropping na values and values where population data is not corrupted
cities_pop_valid = cities_pop.dropna(subset='population')
cities_pop_valid = cities_pop_valid[cities_pop_valid['population'].str.isnumeric()]
cities_pop_valid
```

```python
# Turn data into int, then filtering for cities with more than 100,000 inhabitants
cities_pop_valid = cities_pop_valid.astype({'population': 'int'})
cities_pop_valid_above_100000 = cities_pop_valid[cities_pop_valid['population'] >= 100000]
cities_pop_valid_above_100000
```

```python
# Merge both population and temperature datasets
large_cities_pop_temp = cities_pop_valid_above_100000.merge(cities_temp, left_on='asciiname', right_on='City')
```

```python
# Clean it a little further
large_cities_pop_temp.drop(['asciiname','country code','Continent'], axis=1, inplace=True)
```

```python
# Coldest city with at leas 100,000 inhabitants
large_cities_pop_temp.iloc[large_cities_pop_temp.loc[large_cities_pop_temp['population'] >= 0]['Year'].idxmin()]
```
